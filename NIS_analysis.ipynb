```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Radar NIS plotting

```python
file_radar = open("build/NIS_radar.txt", "r") 
arr_radar = [rad for rad in file_radar.read().split(';')]
radar_values = []
for rad_nis in arr_radar:
    if(rad_nis != ''):
        radar_values.append(float(rad_nis))
plt.figure()
plt.plot(radar_values[1:])
x_95 = [7.815 for _ in radar_values]
x_05 = [0.352 for _ in radar_values]
plt.plot(x_95)
plt.plot(x_05)
plt.show()
```

### Radar NIS calculation

```python
x_95 = 0
x_05 = 0
total = len(radar_values)
for value in radar_values:
    if(value > 0.352):
        x_95 += 1
    if(value > 7.815):
        x_05 += 1
x_95 = x_95 * 100 / total
x_05 = x_05 * 100 / total
print("x_95 = {0}%".format(x_95))
print("x_05 = {0}%".format(x_05))
```
